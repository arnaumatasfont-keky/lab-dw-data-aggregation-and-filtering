```python
df['State'] = df['State'].fillna('Unknown')
#para no perder los datos que no tienen estado, les asignamos el valor 'Unknown'
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
print(df.columns)
```

```python
f_low_claim_yes = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')].copy()
f_low_claim_yes.head()
```

```python
df_yes = df[df['response'] == 'Yes'].copy()
analysis_mean_yes = df_yes.groupby(['policy_type', 'gender'])[['monthly_premium_auto', 'customer_lifetime_value', 'total_claim_amount']].mean()
analysis_mean_yes
```

```python
#Aun que las difrencias entre sexos no son destacables, podemos observar que los hombres tienden a tener polizas más bajas que las mujeres, por lo tanto, son un sector menos provechoso que el femenino. Aun así, en "personal autos" podemos ver como, aun pagando menos, reclaman más. 
```

```python
states_count= df.groupby('state')['customer'].count()
states_count
```

```python
top_states = states_count[states_count > 500]
top_states
```

```python
clv_analysis = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median', 'std'])
clv_analysis
```

```python
clv_analysis = clv_analysis.sort_values(by='median', ascending=False)
print(clv_analysis)
```

```python
#Podemos ver que los valores de min y max son más bajos en M que en F, pero que la median es más alto en M. Para poder entender este dato hemos calculado también la "std" y hemos visto como la dispersión en mujeres es mucho más amplia,. Podemoc conluir que la mayoria de hombre estan más concentrados en una mayor cantidad, pero menos representados en los extremos. Eso nos podría llevar a intentar hacer un marqueting más estándar para hombres y más especifico para mujer, debido a su alta dispersión.
```
